## Experts implementation - dynamic data
This is a example for sending pandas dataframe via the TCP protocol from a client to a server and showing up th data on the fly. Just for bonus task 6. <br>
Adapted from: https://stackoverflow.com/questions/65686506/how-to-make-a-fast-continuous-stream-of-data-with-socket-on-python

```python
from socket import SOCK_DGRAM, SOCK_STREAM, SO_REUSEADDR
from time import sleep
import socket, errno, time, pickle
import pandas as pd
import threading
import os, io
_server = None
_client = None
```

```python
try:
    _server.stopRecieving()
except:
    pass

class server:
    def __init__(self, PORT=9077, MAX_CONNECTIONS=10, BUFF_SIZE=4096):
        self.s = socket.socket(socket.AF_INET, type=SOCK_STREAM)
        self.HOST = socket.gethostbyname(socket.gethostname())
        self.PORT = PORT
        self.MAX_CONNECTIONS = MAX_CONNECTIONS
        self.BUFF_SIZE = BUFF_SIZE
        self.s.bind((self.HOST, self.PORT))
        self.s.listen(self.MAX_CONNECTIONS)
        self.recievingData = False
        
        self.thread = threading.Thread(target=self.recieveData)
        self.thread.start()
        #self.s.setblocking(False)
        self.startRecieving()
        
        print("Starting a server")
        print("IP: " + str(self.HOST))
        print("Port: " + str(self.PORT))

    def startRecieving(self):
        self.recievingData = True
        self.thread = threading.Thread(target=self.recieveData)
        self.thread.start()
        print("Started reading data")
    
    def stopRecieving(self):
        self.recievingData = False
        self.s.close()
        self.thread.join()
        print("Stopped reading data")
        sleep(1)

    def onRecieveData(self, data):
        print("Recieved data!")
        vehicle_id, track_id, modality, modality_precission, df = pd.read_pickle(io.BytesIO(data))
        display(df)

    def receiveDataHelper(self, conn):
        data_size = struct.unpack('>I', conn.recv(4))[0]
        received_payload = b""
        reamining_payload_size = data_size
        while reamining_payload_size != 0:
            received_payload += conn.recv(reamining_payload_size)
            reamining_payload_size = data_size - len(received_payload)
        return received_payload
    
    def recieveData(self):
        self.recievingData = True
        conn, addr = self.s.accept()
        while self.recievingData:
            try:
                data = self.receiveDataHelper(conn)
                if data is not None:
                    self.onRecieveData(data)
            except socket.error as e:
                if e.args[0] == errno.EWOULDBLOCK: 
                    time.sleep(1)           # short delay, no tight loops
                else:
                    print(e)
                    break
_server = server()
```

```python
import struct
class client:
    def __init__(self, HOST=_server.HOST, PORT=_server.PORT):
        self.s = socket.socket(socket.AF_INET, type=SOCK_STREAM)
        self.s.connect((HOST, PORT))
        self.HOST = HOST
        self.PORT = PORT

    def send_message(self, serialized_data):
        self.s.sendall(struct.pack('>I', len(serialized_data)))
        self.s.sendall(serialized_data)
        
_client = client()

from os import walk
for (dirpath, dirnames, filenames) in walk('data/User1_Smartphone/'):
    for file in filenames:
        with open(os.path.join(dirpath, file), 'rb') as f:
            print("Sent:", file)
            _client.send_message(f.read())
            sleep(1)

```
